```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, confusion_matrix, classification_report, roc_auc_score
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming final_merged_with_products_df is your DataFrame
# Select relevant columns
columns_of_interest = [
    'AnnualIncome', 'TotalChildren', 'OrderLineItem', 'OrderQuantity',
    'ProductSize', 'ProductCost', 'ProductPrice'
]

# Filter DataFrame to include only these columns
filtered_df = final_merged_with_products_df[columns_of_interest]

# Check for missing values and handle them
filtered_df = filtered_df.dropna()

# Define the dependent variable (target) and independent variables (predictors)
X = filtered_df.drop(columns='ProductPrice')
y = filtered_df['ProductPrice']

# Convert categorical variables to numeric using one-hot encoding if needed
X_encoded = pd.get_dummies(X, columns=['OrderLineItem', 'ProductSize'])

# Create a binary target based on a threshold for ProductPrice BEFORE splitting the data
threshold = 100  # Example threshold
y_binary = (y > threshold).astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train_binary, y_test_binary = train_test_split(X_encoded, y_binary, test_size=0.2, random_state=42)

# Train a logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train_binary)

# Predict probabilities
y_prob = logreg.predict_proba(X_test)[:, 1]

# Generate ROC curve
fpr, tpr, thresholds = roc_curve(y_test_binary, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

# Confusion Matrix
y_pred = logreg.predict(X_test)
conf_matrix = confusion_matrix(y_test_binary, y_pred)
print('Confusion Matrix:\n', conf_matrix)

# Classification Report
print('Classification Report:\n', classification_report(y_test_binary, y_pred))
```
